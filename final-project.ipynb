## Dataset Preparation

### Importing and Preparing Datasets
Before I can start performing any sort of analysis on the datasets, I will need to import them into the Jupyter notebook working environment. As iterated in the 'Data Collection' section previously, the datasets I will be using are the ["Energy Indicators"](https://unstats.un.org/unsd/envstats/qindicators) dataset and the ["Global health estimates: Leading causes of death"](https://www.who.int/data/gho/data/themes/mortality-and-global-health-estimates/ghe-leading-causes-of-death) dataset, which can both be found in the 'data' folder of this repository. To do import the datasets, I will be using the `pandas` library. The "Energy Indicators" ('energy_indicators.csv') dataset has 5 features: `countryID`, `country_and_area`, `energy_supply` (in petajoules), `energy_supply_per_capita` (gigajoules per capita), and `contribution_of_renewables_to_electricity_production` (%). The "Global health estimates: Leading causes of death" ('death_rates.csv') dataset features including the death rates for the top 10 causes of mortality, aforementioned in this study.

```python
# Importing the pandas library
import pandas as pd
```

After the `pandas` library has been imported, I can now create a DataFrame using the data in the CSV file. The `pandas` library has a myriad of built-in functions that are suitable for data analysis and aggregation.

```python
# Transforming the CSV file to a pandas DataFrame
energy_indicators_df = pd.read_csv("./data/energy_indicators.csv", na_values=False)
death_rates_df = pd.read_csv("./data/death_rates.csv", na_values=False)
```

Now that is all done, there is one more thing that I need to standardize before I am able to cross compare and analylze both of the datasets against each other. In order to join datasets together, I will need to standardize the `country_and_area` feature in the `energy_indicators` DataFrame and the `country_ISO_code` feature from the `death_rates` DataFrame to each country's respective short names. To do this, I will use the [country-converter](https://pypi.org/project/country-converter/#data-sources-and-further-reading) Python package. The first step is to run:

```python
pip install country_converter --upgrade
```

in your working environment terminal to ensure the package is up to date. Next, we can convert the country name columns in both DataFrames and add it as a new column to each respective DataFrame. Referencing the [country-converter](https://pypi.org/project/country-converter/#data-sources-and-further-reading) documentation, we can do this in a couple lines of code.

```python
# Importing the country-converter package to use in code
import country_converter as coco

# Saving the existing names columns as a list in pandas
countries_ei = energy_indicators_df["country_and_area"]
countries_dr = death_rates_df["country_ISO_code"]

# Converting the names of the variation of country names using the 'country-converter' package
short_names_ei = coco.convert(names=countries_ei, to='name_short')
short_names_dr = coco.convert(names=countries_dr, to='name_short')

# Adding the short names of countries as a new column into each respective DataFrame
energy_indicators_df.insert(2, "country_short_name", short_names_ei)
death_rates_df.insert(1, "country_short_name", short_names_dr)

# Printing the length of country names for each dataset
print(len(short_names_ei))
print(len(short_names_dr))
```

The result of the previous cell's output shows us that there are more countries in the `energy_indicators` DataFrame than the `death_rates` DataFrame; this will be useful information as we move forward in joining the two DataFrames. Next, we will check the resulting output of each DataFrame after adding a new column of short names to each DataFrame.

```python
print("Energy Indicators: \n")
display(energy_indicators_df)
print("Death Rates: \n")
display(death_rates_df)
```

With the previous output, we see that the new columns with the standardized short names looks good. However, we see that in the `death_rates` DataFrame, there are multiple additional "unnamed" columns with NaN values. This would cause some errors when we do our data analysis so we will modify the DataFrame to include the columns that only contain non-NaN values.

```python
# Selecting only the first 12 columns that include non-NaN values and the data that we will need
death_rates_df = death_rates_df.iloc[:, 0:13]
print("Modified Death Rates DataFrame : \n")
display(death_rates_df)
```

This DataFrame looks much better. Now we will join the two DataFrames in order to align features by countries. We can do this by using the `merge` function in Pandas. Recall that earlier there were more countries in the `energy_indicators` DataFrame than in the `death_rates` one; after merging/joining our two DataFrames together on country name, the countries that are not contained in `death_rates` will be omitted from the resulting joined DataFrame.

```python
# Joining the two DataFrames with their common 'country_short_name' column
merged_df = death_rates_df.merge(energy_indicators_df, left_on='country_short_name', right_on='country_short_name')
print('Merged:')
display(merged_df)

# Print out the length of the merged DataFrame
print("No. of rows: " + str(len(merged_df)))
```

Notice that number of countries in the merged DataFrame has the same number of columns, 183, as the original `death_rates` DataFrame. Before we can start looking into deeper statistical analysis, one more thing to ensure is that the object types for each numerical data column is the same. This ensures less hiccups down the road during data analyis. To check the data types of values in each column, we will use a loop to iterate across all columns in one row and print out the column name as well as the data type for the data value.

```python
df_cols = merged_df.columns
for col in df_cols:
    print(col)
    print(type(merged_df[col][1]))
```

We can see that `country_ISO_code`, `country_short_name`, and `country_and_area` contain string values, which is accurate. However, we see inconsistencies among the columns containing statistical numerical data; the `population` column has type int, `energy_supply_per_capita` has type int, and `contribution_of_renewables_to_energy_production` has type string, while other statistical numercial data columns have type float. We will need to modify the inconsistencies by using the `astype()` method that is built-in to pandas.

```python
merged_df['population'] = merged_df['population'].astype(float)
merged_df['energy_supply_per_capita'] = merged_df['energy_supply_per_capita'].astype(float)
merged_df['contribution_of_renewables_to_energy_production'] = merged_df['contribution_of_renewables_to_energy_production'].astype(float)
```

## Findings

### (1) Renewable Energy Production and Mortality Rates Across Countries
This section will investigate the amount of correlation between the renewable energy production of a country and how that relates to the mortality rate for each of the top 10 causes of death. To start, we will sort the rows by the column `contribution_of_renewables_to_energy_production`, which yields percentages, in ascending order and save it to the variable `ordered_df`. Note: the death rates for the causes of death in the DataFrame is the number of deaths per 100,000 persons.

```python
# Sorting and displaying the merged DataFrame by the 'contribution_of_renewables_to_energy_production' column in ascending order
ordered_df = merged_df.sort_values('contribution_of_renewables_to_energy_production', ascending=True)
display(ordered_df)
```

Although sorting the rows in ascending order by `contribution_of_renewables_to_energy_production` is not necessary if we are to use the pandas library, it is interesting to viusalize the tabular data in this way. First, we'll take a look at Ischaemic Heart Disease as cause of mortality and its relationship to renewable energy production rate by country. We will first need to isolate for specific columns: `contribution_of_renewables_to_energy_production` and `ischaemic_heart_disease`

```python
heart_disease_df = ordered_df.loc[:, ['contribution_of_renewables_to_energy_production', 'ischaemic_heart_disease']]
display(heart_disease_df)
```

Now, we can use the built-in `corr` method on the smaller `heart_disease_df` DataFrame to find Pearson's correlation coefficient between renewable energy production (%) and the death rate by 100,000 for ischaemic heart disease and save it to the variable `r_ischaemic_heart_disease`.

```python
# Creating a correlation matrix using the 'corr' method on the 'heart_disease_df'
heart_disease_corr_matrix = heart_disease_df.corr(method='pearson')
r_ischaemic_heart_disease = heart_disease_corr_matrix.iloc[0][1]
```

From simple numerical analysis, we see that there is a very slight negative correlation between renewable energy production and death from ischaemic heart disease with the correlation coefficient being -0.14, a number that is negative and very close to 0. Now, we will display this information graphically using a scatterplot that displays "Contribution of Renewable Energy Production (%)" on the x-axis plotted against the "Death Rate (per 100,000 persons)". For plots that are non-geographical plots, we will be using the [Seaborn library](https://seaborn.pydata.org/). We need to import, then call `set` on the library.

```python
import seaborn as sns
sns.set()
```

Now, we can create our scatterplot calling `scatterplot` on our imported seaborn library and specifying the dataset, the x, the y; we also want to label the plot using matplotlib.pyplot, which we imported earlier. 

```python
# Creating the scatterplot for ischaemic heart disease death rate
sns.scatterplot(data=heart_disease_df, x='contribution_of_renewables_to_energy_production', y='ischaemic_heart_disease')

# Labelling the plot for ischaemic heart disease death rate
plt.xlabel('Contribution of Renewable Energy Production (%)')
plt.ylabel('IHD Death Rate (per 100,000 persons)')
plt.title("Relationship between renewable energy production and ischaemic heart disease (IHD) deaths (r=" + str(round(r_ischaemic_heart_disease, 2)) + ")")
```

### (2) Average Mortality Rates for the Top 10 Causes of Death 
Now, we will find the average mortality rates for each cause of death. In order to do this, we will use the `ordered_df` because we had already done the work of aggregating our datasets together. This dataset offers us the comprehensive dataset we will need to complete this section of analysis. First we will break down the DataFrame into a smaller one that includes only the death rates of the causes of death, country name, country ISO code, and contribution to renewable energy production.

```python
heatmap_df = ordered_df.loc[:, ['country_ISO_code', 'country_short_name', 'ischaemic_heart_disease', 'stroke', 'chronic_obstructive_pulmonary_disease', 'lower_respiratory_infections', 'neonatal_conditions', 'trachae_lung_bronchus_cancer', 'alzheimers_disease_dementia', 'diahorreal_diseases', 'diabetes_mellitus', 'kidney_diseases', 'contribution_of_renewables_to_energy_production']]
```

```python
display(heatmap_df)
```

On a separate note, we will need to calculate the averages for each of the death rates for each disease. We can do this by calling the built-in pandas `.mean()` method on each of the columns that contains data for disease death rates. Creating a function to do this will save code repetition. We can apply this function to each of the columns we would like to calculate the mean for. We will save these average values as a list to use in the future.

```python
# Finds the average value in a specified column in the DataFrame 'heatmap_df'; takes a string of a column in 'heatmap_df' and returns a float that is rounded to 2 decimal places.
def find_avg_death_rate(cause_of_death):
    return round(heatmap_df[cause_of_death].mean(), 2)

# Getting a list of the column names for the causes of death
causes_of_death = heatmap_df.columns.values.tolist()[2: 12]

# Applying 'find_avg_death_rate' to each of the causes of death
death_rate_averages = []
for cause in causes_of_death:
    death_rate_averages.append(find_avg_death_rate(cause))
    
print(death_rate_averages)
print(causes_of_death)
```

Before we can create geographical plots and overlay our data onto them, we will need to join the `countries` dataframe we prepared earlier with our `heatmap_df`. If we `display(countries)` we will see that it has 3 columns, one of which, `ISO_A3`, can be joined with the `country_ISO_code` column in `heatmap_df`. 

```python
display(countries)
```

We will modify the column name `ISO_A3` to align with the column name in `heatmap_df`:

```python
countries = countries.rename(columns={"ISO_A3":"country_ISO_code"})
display(countries)
```

Now to join `countries` and `heatmap_df` on country ISO code to create a new DataFrame that is of type GeoDatFrame:

```python
countries_geo = countries.merge(heatmap_df, how='right', on='country_ISO_code')
print(type(countries_geo))
display(countries_geo)
```

Now that we have a list of the average rates for each cause of death, we can create 10 different plots and use a the `TwoSlopeNorm` package to clearly show which countries fall above or below the average death rate value for each cause of death. We will use the geopandas library to create these charts. To do this we will first import the package into our working environment. We will also need numpy to work with specific numerical terms:

```python
from matplotlib.colors import TwoSlopeNorm
import numpy as np
```

We will perform color normalizing in order to emphasize the average value for each cause of death for each map that we create.:

```python
# Normalize color
vmin, vcenter, vmax = min(countries_geo['ischaemic_heart_disease']), death_rate_averages[0], max(countries_geo['ischaemic_heart_disease'])
norm = TwoSlopeNorm(vmin=vmin, vcenter=vcenter, vmax=vmax)

# Create a normalized colorbar
cmap = 'PiYG'
cbar = plt.cm.ScalarMappable(norm=norm, cmap=cmap)

# Creating subplots
fig2, ax2 = plt.subplots(figsize=(20, 6))

# Plotting and adding color and labels
countries_geo.plot(column='ischaemic_heart_disease', cmap=cmap, norm=norm, ax=ax2)
fig2.colorbar(cbar, ax=ax2)
ax2.axis('off')
ax2.set_title("Ischaemic heart disease death rates across countries (average = " + str(death_rate_averages[0]) + ")")
```

From the result graph of the previous cell's output we can see that the average death rate for ischaemic heart disease is white and the darker a hue is corresponding to the distant from the average mortality rate. Now, before we graph all of the other plots for the other nine causes of death, we'll graph the contribution percentage to renewable energy on a geographical plot so that we can see how death rates for each death cause relates to the percentage of renewable energy by country. We will also save this map into the 'plots' directory for future reference under `map_for_renewable_energy_contribution.png`

```python
# Normalize color
vmin, vcenter, vmax = 0.0, 50.0, 100.0
norm = TwoSlopeNorm(vmin=vmin, vcenter=vcenter, vmax=vmax)

# Create a normalized colorbar
cmap = 'RdYlBu'
cbar = plt.cm.ScalarMappable(norm=norm, cmap=cmap)

# Creating subplots
fig3, ax3 = plt.subplots(figsize=(20, 6))

# Plotting and adding color and labels
countries_geo.plot(column='contribution_of_renewables_to_energy_production', cmap=cmap, norm=norm, ax=ax3)
fig3.colorbar(cbar, ax=ax3)
ax3.set_title("Contribution of renewable energy to total energy production (%)")
ax3.axis('off')

# Saving the plot into the 'plots' directory
plt.savefig('./plots/map_for_renewable_energy_contribution.png', bbox_inches='tight')
```

Now, we can create a subplot figure that shows the 'Contribution of renewable energy to total energy production (%)' and all of the figures that shows how countries compare with each other against the respective average death rate for each of the top 10 leading causes of death. To do this, we will create a function that uses loops and uses the list `causes_of_death` that we created earlier and `death_rate_averages` for each cause of death. Additionally, we will create a list of names that can be used to title each of the plots we create in the loop. Lastly, we will save this figure with all the maps into the 'plots' directory under `maps_for_death_rates.png` for reference during result analysis.

```python
# Creating subplots for the figure
fig4, ax4 = plt.subplots(11, figsize=(20, 70))

## Map plot that shows the contribution of renewable energy to total energy production for reference against the rest of the plots

# Normalize color
vmin, vcenter, vmax = 0.0, 50.0, 100.0
norm = TwoSlopeNorm(vmin=vmin, vcenter=vcenter, vmax=vmax)

# Create a normalized colorbar
cmap = 'RdYlBu'
cbar = plt.cm.ScalarMappable(norm=norm, cmap=cmap)

# Plotting and adding color and labels, removing axis
countries_geo.plot(column='contribution_of_renewables_to_energy_production', cmap=cmap, norm=norm, ax=ax4[0])
fig4.colorbar(cbar, ax=ax4[0])
ax4[0].axis('off')
ax4[0].set_title("Contribution of renewable energy to total energy production (%)")

###########################
# Creating a list for each cause of death that can be iterated through to title each plot
death_cause_title_names = ['Ischaemic heart disease', 'Stroke', 'Chronic obstructive pulmonary disease', 'Lower respiratory infections', 'Neonatal conditions', 
                           'Trachea, bronchus, lung cancer', "Alzheimer's disease and other dementia",
                           'Diarrhoeal diseases', 'Diabetes mellitus', 'Kidney diseases']

###########################
# Generalizing the method of plotting each map to reduce code redundancy and for efficiency

def graph_avg_death_rates(causes_of_death):
    count = 0
    count_add_1 = 1
    for cause in causes_of_death:
        
        vmin, vcenter, vmax = min(countries_geo[cause]), death_rate_averages[count], max(countries_geo[cause])
        norm = TwoSlopeNorm(vmin=vmin, vcenter=vcenter, vmax=vmax)
        
        cmap = 'PiYG'
        cbar = plt.cm.ScalarMappable(norm=norm, cmap=cmap)
        
        countries_geo.plot(column=cause, cmap=cmap, norm=norm, ax=ax4[count_add_1])
        fig4.colorbar(cbar, ax=ax4[count_add_1])

        ax4[count_add_1].axis('off')
        ax4[count_add_1].set_title(death_cause_title_names[count] + " death rates across countries (average = " + str(death_rate_averages[count]) + ")")

        count = count + 1
        count_add_1 = count_add_1 + 1
        
    
###########################
# Calling the function to graph all the plots for each cause of death
graph_avg_death_rates(causes_of_death)

# Saving the figure into the 'plots' directory
plt.savefig('./plots/maps_for_death_rates.png', bbox_inches='tight')
```
